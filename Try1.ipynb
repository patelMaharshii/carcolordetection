```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from scipy.io import loadmat
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import albumentations as A
from albumentations.pytorch import ToTensorV2
import time
```

```python
base_path = "./arxiv_data/stanford_cars"
train_dir = os.path.join(base_path, "cars_train")
test_dir = os.path.join(base_path, "cars_test")
train_annos = os.path.join(base_path, "devkit", "cars_train_annos.mat")
test_annos = os.path.join(base_path, "", "cars_test_annos_withlabels.mat")
```

### Architechture

```python
# Lightweight Model Architecture
class CarSegmenter(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = nn.Sequential(
            nn.Conv2d(3, 16, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(16, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(64, 32, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(32, 16, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 1, 1)
        )

    def forward(self, x):
        x = self.backbone(x)
        return self.decoder(x)
```

### Dataset Loading & Augmentation

```python
# Dataset Class with Augmentation
class StanfordCarsDataset(Dataset):
    def __init__(self, root_dir, annotations_path, transform=None):
        self.root_dir = root_dir
        self.annotations = loadmat(annotations_path)['annotations'][0]
        self.transform = transform
        
        # Extract all items once during init
        self.samples = []
        for anno in self.annotations:
            img_name = anno[-1].item()
            bbox = (anno[0].item(), anno[1].item(), 
                    anno[2].item(), anno[3].item())
            class_id = anno[4].item() if 'train' in annotations_path else anno[5].item()
            self.samples.append((img_name, bbox, class_id))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_name, bbox, class_id = self.samples[idx]
        img_path = os.path.join(self.root_dir, img_name)
        
        image = np.array(Image.open(img_path).convert("RGB"))
        cropped = image[bbox[1]:bbox[3], bbox[0]:bbox[2]]
        
        # Generate pseudo-mask (replace with real masks if available)
        mask = np.zeros(cropped.shape[:2], dtype=np.float32)
        h, w = mask.shape
        mask[h//4:-h//4, w//4:-w//4] = 1  # Central rectangle
        
        if self.transform:
            augmented = self.transform(image=cropped, mask=mask)
            return augmented['image'], augmented['mask'].float()
            
        return transforms.ToTensor()(cropped), torch.from_numpy(mask).float()
```

```python
# Augmentations
train_transform = A.Compose([
    A.Resize(256, 256),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2(),
])
```

```python
# Initialize Datasets and DataLoaders
train_dataset = StanfordCarsDataset(
    root_dir=train_dir,
    annotations_path=train_annos,
    transform=train_transform
)

test_dataset = StanfordCarsDataset(
    root_dir=test_dir,
    annotations_path=test_annos,
    transform=train_transform  # Same transform for simplicity
)
```

```python
BATCH_SIZE = 16
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE)
```

### Training Setup

```python
device = torch.device('mps' if torch.backends.mps.is_available() else 'cpu')
model = CarSegmenter().to(device)
optimizer = optim.AdamW(model.parameters(), lr=1e-3)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=2)
```

```python
def dice_loss(pred, target):
    smooth = 1.
    pred = torch.sigmoid(pred)
    intersection = (pred * target).sum()
    return 1 - (2. * intersection + smooth) / (pred.sum() + target.sum() + smooth)
```

```python
def train_epoch(model, loader, optimizer):
    model.train()
    total_loss = 0
    for images, masks in loader:
        images, masks = images.to(device), masks.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = dice_loss(outputs.squeeze(1), masks)
        
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    return total_loss / len(loader)
```

### Training Model

```python
start = time.time()

# Full Training Loop
for epoch in range(15):
    train_loss = train_epoch(model, train_loader, optimizer)
    print(f"Epoch {epoch+1} | Loss: {train_loss:.4f}")
    scheduler.step(train_loss)
    
    # Save checkpoint
    if (epoch+1) % 5 == 0:
        torch.save(model.state_dict(), f"car_segmenter_epoch{epoch+1}.pth")
        
stop = time.time()
time_elapsed = stop - start

print(f"1 epoch took: {time_elapsed} seconds")
```

### Testing Model

```python
def remove_outliers_iqr(data):
    Q1 = np.percentile(data, 25, axis=0)
    Q3 = np.percentile(data, 75, axis=0)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    mask = np.all((data >= lower_bound) & (data <= upper_bound), axis=1)
    return data[mask]
```

```python
# def predict_car_color(image_path, model, bbox, device='cpu'):
    
#     image = Image.open(image_path).convert("RGB")
    
#     x1, y1, x2, y2 = bbox
#     image = image.crop((x1, y1, x2, y2))
#     original_size = image.size
    
#     preprocess = transforms.Compose([
#         transforms.Resize(256),  # Resize shorter side to 520
#         transforms.ToTensor(),
#         transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
#     ])
    
#     input_tensor = preprocess(image).unsqueeze(0).to(device)
    
#     with torch.no_grad():
#         output = model(input_tensor)
        
#     # Generate mask
#     mask = torch.sigmoid(output.squeeze()) > 0.5 
    
#     # Resize mask to original cropped size
#     mask = transforms.Resize(original_size[::-1],  # (height, width)
#                            interpolation=transforms.InterpolationMode.NEAREST)(mask.float())
    
#     # Convert to numpy
#     mask = mask.squeeze().cpu().numpy().astype(bool)
    
#     # Extract and calculate median color
#     car_pixels = np.array(image)[mask]
#     return np.median(car_pixels, axis=0) if car_pixels.size > 0 else None

def predict_car_color(image_path, model, bbox, device='cpu'):
    # Load and crop image
    image = Image.open(image_path).convert("RGB")
    x1, y1, x2, y2 = bbox
    cropped_img = image.crop((x1, y1, x2, y2))
    original_size = cropped_img.size  # (width, height)
    
    # Preprocessing pipeline
    preprocess = transforms.Compose([
        transforms.Resize(256),  # Match model's expected input size
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    
    # Process image
    input_tensor = preprocess(cropped_img).unsqueeze(0).to(device)
    
    with torch.no_grad():
        output = model(input_tensor)
        
    # Handle different model architectures
    if isinstance(output, dict):
        output = output['out']  # For DeepLabV3-style models
    output = output.squeeze(1)  # Remove channel dimension if needed
    
    # Generate mask (ensure proper dimensions)
    mask = torch.sigmoid(output).squeeze().cpu().numpy() > 0.5
    
    # Convert to PIL Image for proper resizing
    mask_img = Image.fromarray((mask * 255).astype(np.uint8))
    
    # Resize to original cropped dimensions using NEAREST neighbor
    mask_img = mask_img.resize(original_size, resample=Image.NEAREST)
    mask = np.array(mask_img).astype(bool)
    
    # Extract and calculate median color
    car_pixels = np.array(cropped_img)[mask]
    filtered_pixels = remove_outliers_iqr(car_pixels)
    return np.mean(filtered_pixels, axis=0) if car_pixels.size > 0 else None
```

```python
def visualize_prediction(image_path, model, device='cpu'):
    # Load and process image
    image = Image.open(image_path).convert("RGB")
    
    # Create dataset instance without transforms
    dataset = StanfordCarsDataset(
        root_dir=os.path.dirname(image_path),
        annotations_path=test_annos,
        transform=None
    )
    
    # Find matching annotation using filename
    img_name = os.path.basename(image_path)
    idx = [i for i, sample in enumerate(dataset.samples) if sample[0] == img_name][0]
    
    # Get stored annotation (not using __getitem__)
    img_name, bbox, class_id = dataset.samples[idx]
    
    # Get prediction
    median_color = predict_car_color(image_path, model, bbox)
    
    # Create plot
    plt.figure(figsize=(15, 6))
    
    # Original image with bounding box
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    rect = plt.Rectangle(
        (bbox[0], bbox[1]), 
        bbox[2]-bbox[0], 
        bbox[3]-bbox[1],
        linewidth=2, 
        edgecolor='r', 
        facecolor='none'
    )
    plt.gca().add_patch(rect)
    plt.title(f"Original Image: {img_name}")
    
    # Median color swatch
    plt.subplot(1, 2, 2)
    if median_color is not None:
        plt.imshow([[median_color.astype(int)]])
        plt.title(f"Median Color: {median_color.astype(int)}")
    else:
        plt.text(0.5, 0.5, "No car pixels detected", ha='center')
        plt.title("Color Detection Failed")
    
    plt.tight_layout()
    plt.show()
    
    return dataset
```
